# 03 — Evaluation & Analysis

Compute all system + per-model metrics, compare the three systems (Single-LLM, Voting, Full Debate), and render the 7 required plots.

> Run `python src/pipeline.py` first so that `data/results/final_answers.json` exists.

```python
import sys, json
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
from src import config, evaluator

records = evaluator.load_records(config.RESULTS_DIR)
details = evaluator.load_details(config.RESULTS_DIR)
metrics = evaluator.compute_metrics(records, details)
evaluator.print_summary(metrics)
```

## Generate all 7 plots and display them inline

```python
for fn in evaluator.ALL_PLOTS:
    fn(metrics, config.PLOTS_DIR)
    print('generated', fn.__name__)
```

```python
from IPython.display import Image, display
plots = ['overall_accuracy', 'per_category_accuracy', 'improvement_rate',
         'consensus_rate', 'judge_accuracy', 'per_model_accuracy', 'confidence_calibration']
for name in plots:
    print(name)
    display(Image(str(config.PLOTS_DIR / f'{name}.png')))
```
