# 01 — Dataset Construction

Build and inspect the 25-problem benchmark with verifiable ground-truth answers.
Each problem has: `id`, `category`, `problem`, `correct_answer`, `difficulty`, `source_notes`.

```python
import json, sys
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
problems = json.loads((ROOT / 'data' / 'problems.json').read_text(encoding='utf-8'))
print(f'Loaded {len(problems)} problems')
df = pd.DataFrame(problems)
df.head()
```

## Schema validation
Every problem must contain all required fields and a non-empty ground-truth answer.

```python
REQUIRED = {'id', 'category', 'problem', 'correct_answer', 'difficulty', 'source_notes'}
assert len(problems) == 25, f'Expected 25 problems, got {len(problems)}'
ids = set()
for p in problems:
    assert REQUIRED.issubset(p.keys()), f'{p.get("id")} missing fields'
    assert str(p['correct_answer']).strip(), f'{p["id"]} has empty answer'
    assert p['id'] not in ids, f'duplicate id {p["id"]}'
    ids.add(p['id'])
print('All 25 problems pass schema validation. Unique ids:', len(ids))
```

## Distribution by category and difficulty

```python
print('By category:')
print(df['category'].value_counts())
print('\nBy difficulty:')
print(df['difficulty'].value_counts())
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['category'].value_counts().plot(kind='bar', ax=axes[0], color='#4C8BE2', title='Problems per category')
df['difficulty'].value_counts().plot(kind='bar', ax=axes[1], color='#2BB673', title='Problems per difficulty')
plt.tight_layout(); plt.show()
```

## Full problem listing

```python
for p in problems:
    print(f"[{p['id']}] ({p['category']}, {p['difficulty']})")
    print(f"  Q: {p['problem']}")
    print(f"  A: {p['correct_answer']}\n")
```
